**1) Realiza un programa que lea tres números por teclado y permita evaluar lo siguiente:** 
* Si los números están en orden ascendente
* Si los números no están en orden ascendente
* Si hay un error debido a que el primer número introducido es 0

__No uses listas, eso dejalo para el siguiente ejercicio__

````python
#Pedimos al usuario numeros#
num1 = int(input("Introduce un número: "))
num2 = int(input("Introduce un número: "))
num3 = int(input("Introduce un número: "))
#Usando la sentencia if, elif y else recorremos los numeros introducidos por el usuario para determinar si es asc-desc-o invalido#
if num1 == 0:
    print("Error el primero número no puede ser 0")
elif num1 < num2 and num2 < num3:
    print("Números en orden ascendente")
else:
    print("Números en orden descendente")
````

**3) Realiza un programa que realice el sumatorio de los primeros N números, empezando en 0, y siendo N el número límite que el usuario proporciona por teclado:** 
* Ejemplo de número límite introducido por el usuario: 5
* sumatorio = 0 + 1 + 2 + 3 + 4 (sumamos los 5 primeros números)

````python
# Pedimos al usuario que introduzca un número límite#
nNumeros = int(input("Introduce un número: "))
sumatorio = 0 #Inicializar la variable para el sumatorio#
for i in range(nNumeros): #Bucle for usando range para realizar la suma de los N primeros numeros#
    sumatorio += i
    print()
print("El sumatorio de los primeros", nNumeros, "números es:", sumatorio)
````

**4) Realiza un programa que lea letras y cuente con una variable contador las letras "a" que se introducen. Para salir del programa, introducir el carácter ".". Al finalizar mostrar el número de veces que se ha pulsado la letra "a".**

````python
cadena = "Hola Mundo"
contador = 0
for indice, caracter in enumerate(cadena):
    if caracter == 'A' or caracter == 'a':
        print(caracter, "en la posición", indice)
        contador += 1
print("Numero de veces que aparece la letra 'a':", contador)
````

**5) Realiza el sumatorio de los números pares comprendidos entre 1000 y 73 (en este orden, yendo hacia atrás). Implementa esta solución con un bucle for o while:**

````python
suma = 0 #inicializo la variable

for numero in range(1000, 72, -1):
    #print(i)
    if numero % 2 == 0: #compruebo si el numero es par
        suma += numero
print("la suma de los numeros pares comprendidos entre 1000 y 73 es: ", suma)
````

**6) Realiza un programa que solicite al usuario un número que representa el total de números que se van a introducir. Posteriormente leer todos esos números, meterlos en una lista y realizar:**
* La ordenación de la lista de menor a mayor
* Indicar el promedio de la lista
* Indicar el valor máximo y el valor mínimo y en que posiciones se encuentran

Ejemplo de salida por pantalla del ejercicio:


````python
num = int(input("Inroduzca el numero total de numeros: "))
listaNum = []
# Leer los números y agregarlos a la lista
for i in range(num):
    numero = int(input("ingrese un valor entero: "))
    listaNum.append(numero)
listaOrdenada = sorted(listaNum)    
print("La lista de los numeros introducidos es: ", listaNum)
print("La lista de los numeros ordenados es: ", listaOrdenada)
# Calcular el promedio de los valores en listaNum
promedioLista = sum(listaNum) / len(listaNum)
print(promedioLista)

# Encontrar el valor máximo
valorMax = max(listaNum)
indiceMax = []

# Encontrar el valor mínimo
valorMin = min(listaNum)
indiceMin = []

# Buscar las posiciones del valor máximo y mínimo
for i, num in enumerate(listaOrdenada):
    if num == valorMax:
        indiceMax.append(i)
    if num == valorMin:
        indiceMin.append(i)

print("Valor máximo:", valorMax, "en las posiciones:", indiceMax)
print("Valor mínimo:", valorMin, "en las posiciones:", indiceMin)
````

**7) Crea una lista de palabras, recorre la lista y muestra cada palabra junto con su longitud. Al final, indicar cual fue la palabra con más caracteres.**

````python
listaPalabras = ["Java", "Python", "JavaScript", "Kotlin"]
#inicializo las variables donde voy a almacenar la palabra mas larga y su longitud
palabraLarga = " "
longitudPalabra = 0
#bucle for para recorrer la lista de palabras y su longitud
for palabra in listaPalabras:
    longitud = len(palabra)
    print(palabra, longitud)
#sentencia if para determinar cual es la palabra mas larga y su longitud
    if longitud > longitudPalabra:
        longitudPalabra = longitud
        palabraLarga = palabra
 
print("La palabra mas larga es:", palabraLarga,"y su longitud es:", longitudPalabra)
````

**8) Crea un menú con esta forma:**
```python
MENU COMPRA-VENTA
    1. Comprar
    2. Vender
    3. Alquilar
    4. Mostrar listas de datos
    5. Salir
    Introduzca que desea hacer(1-4): 
```
* Si se selecciona la opción 1 preguntaremos al usuario que quiere comprar y después de que el usuario lo haya introducido se guardará en una lista destinada a las compras y se volverá al menú principal.
* Si se selecciona la opción 2 preguntaremos al usuario que quiere vender y después de que el usuario lo haya introducido se guardará en una lista destinada a las ventas y se volverá al menú principal.
* Si se selecciona la opción 3 preguntaremos al usuario que quiere alquilar y después de que el usuario lo haya introducido se guardará en una lista destinada a los alquileres y se volverá al menú principal.
* Si se selecciona la opción 4 se mostrarán todas las listas.
* Si se selecciona la opción 5 saldremos del bucle.

````python
print("Bienvenido a la tienda online")
#inicio varibales para almacenar despues los datos
comprar = []
vender = []
alquilar = []
#inicio un bucle infinito que se ejecutará hasta que el usuario elija la opción de salida 
while(True):
    print("\nMENU COMPRA-VENTA")
    print("1. Comprar")
    print("2. Vender")
    print("3. Alquilar")
    print("4. Mostrar listas de datos")
    print("5. Salir")
    opcion = input("Introduzca lo que desea hacer (1-4): ")
    #Imito un switch para pedir al usuario que opcion quiere
    if opcion == '1':
        producto = input("¿que desea comprar?: ")
        comprar.append(producto)
        print("compra registrada")
    elif opcion == '2':
        producto = input("¿Que desea vender?: ")
        vender.append(producto)
        print("Venta registrada")
    elif opcion =='3':
        producto = input("¿Que desea alquilar?: ")
        alquilar.append(producto)
        print("Alquiler registrado")
    elif opcion =='4':
        print("\nListado de compras:",comprar)
        print("Listado de ventas:", vender)
        print("Listado de alquileres:", alquilar)
    elif opcion =='5':
        print("Saliendo del programa...")
        break
    else:
        print("Comando desconocido, vuelve a intentarlo")
````

**9) Realiza un programa que realice las siguientes tareas:**
* __Paso 1__) solicite al usuario una contraseña y la guarde en una variable 
* __Paso 2__) pida al usuario introducir la contraseña para loguearse
    * si la contraseña es correcta mostramos "__Login correcto__"
    * si la contraseña no es correcta se le tendrá que mostrar un mensaje de "__Login fallido. Intento 1__", donde 1 será una variable contador que irá incrementandose hasta llegar a __5__. En ese punto __reiniciamos el programa y volveremos al Paso 1__

````python
contraseña = input("Introduce una contraseña: ")
# Bucle para solicitar la contraseña y permitir intentos
while True:
    intentos = 0
    while intentos < 5:
        keyUsuario = input("Introduce la contraseña para iniciar sesión: ")
        # Comprobar si la contraseña ingresada coincide con la contraseña guardada
        if keyUsuario == contraseña:
            print("Login correcto")
            break  # Salir del bucle de intentos si la contraseña es correcta
        else:
            intentos += 1
            print("Login fallido", "Intento:", intentos)

    # Si el usuario agota los 5 intentos, reiniciar automáticamente
    if intentos == 5:
        print("Has agotado los 5 intentos. Reiniciando el programa...")
    else:
        break  # Salir del bucle principal si el inicio de sesión es exitoso

print("Programa terminado.")
````
